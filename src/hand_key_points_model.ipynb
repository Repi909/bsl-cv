```python
from ultralytics import YOLO

# Load a model
model = YOLO("yolo11s-pose.pt")  # load a pretrained model (recommended for training)

# Train the model
results = model.train(
    data="hand-keypoints.yaml",
    epochs=100,
    imgsz=640,        # faster
    device="cuda:0",
    workers=12,
    cos_lr=True
)
```
