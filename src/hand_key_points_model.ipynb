```python
from ultralytics import YOLO

# Load a model
model = YOLO("yolo11n-pose.pt")  # load a pretrained model (recommended for training)

# Train the model
results = model.train(
    data="hand-keypoints.yaml",
    epochs=100,
    imgsz=512,        # faster
    batch=16,         # larger batch
    device="cuda:0",
    workers=12,
    half=True,        # FP16 mixed precision
    cache=True,
)
```
